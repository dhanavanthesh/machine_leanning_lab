# Principal Component Analysis (PCA) on Iris Dataset

This program implements PCA to reduce the dimensionality of the Iris dataset from 4 features to 2 principal components.

## What is PCA?
- PCA is a technique for dimensionality reduction
- It transforms high-dimensional data into lower dimensions while preserving as much variance as possible
- Helps in visualization and can improve computational efficiency

## Code Explanation

This code performs PCA on the Iris dataset in a few simple steps:

1. **Import Libraries**:
   - `numpy` for numerical operations
   - `pandas` for data manipulation
   - `sklearn.datasets` to load the Iris dataset
   - `sklearn.decomposition.PCA` for dimensionality reduction
   - `matplotlib.pyplot` for visualization

2. **Load and Prepare Data**:
   - Load the Iris dataset which has 4 features: sepal length, sepal width, petal length, and petal width
   - Store features in `X` and target classes in `y`

3. **Apply PCA**:
   - Create a PCA model with 2 components (reducing from 4 dimensions to 2)
   - Transform the original data using `fit_transform()` into a new 2D space
   - The result `X_pca` contains the coordinates in the new space

4. **Visualize Results**:
   - Create a scatter plot showing each class with a different color
   - The x-axis shows Principal Component 1 (which captures most variance)
   - The y-axis shows Principal Component 2 (which captures second most variance)

This visualization helps us see how well the three Iris species can be separated using just two dimensions instead of the original four features.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Iris dataset and prepare data
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the results
plt.figure(figsize=(8, 6))
for i, label in enumerate(iris.target_names):
    plt.scatter(
        X_pca[y == i, 0], X_pca[y == i, 1],
        label=label
    )

plt.title('PCA on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
