## Understanding Feature Relationships

In this program, we'll explore relationships between different features in the California Housing dataset through:

- **Correlation Matrix**: A table showing correlation coefficients between variables
  - Values close to 1: Strong positive correlation
  - Values close to -1: Strong negative correlation
  - Values close to 0: Little to no correlation

- **Heatmap**: Visual representation of the correlation matrix where colors indicate correlation strength

- **Pair Plot**: Grid of scatterplots showing relationships between pairs of variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Load the California Housing Dataset
housing = fetch_california_housing(as_frame=True)
df = housing.frame

# Compute and visualize correlation matrix
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of California Housing Features')
plt.show()

# Create pair plot (sample for better performance)
sample_df = df.sample(1000) if len(df) > 1000 else df
sns.pairplot(sample_df, height=2.5, plot_kws={'alpha': 0.6})
plt.suptitle('Pair Plot of California Housing Features', y=1.02)
plt.show()
```
